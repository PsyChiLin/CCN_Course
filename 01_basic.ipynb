#### Class

```python
class Man:
    def __init__(self, name):
        self.name = name
        print("Initilized")
        
    def hello(self):
        print("Hello "+ self.name + "!")
    
    def goodbye(self):
        print("Good-bye " + self.name + "!")
```

```python
m = Man("MJ")
```

```python
m.hello()
```

```python
m.goodbye()
```
